```python
# Import necessary dependencies 
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
import random 
from scipy.stats import sem
```

```python
#Create variable to establish csv path and merge on common data set to create comprehensive dataframe 
tumor_trial_df = pd.read_csv("../Resources/clinicaltrial_data.csv")

mouse_drug_data_df = pd.read_csv("../Resources/mouse_drug_data.csv")

drug_trial_complete = pd.merge(tumor_trial_df, mouse_drug_data_df, how="left", on=["Mouse ID", "Mouse ID"])

drug_trial_complete.head()
```

```python
#Deleted column values that I do not need to have a more focused df 
del drug_trial_complete["Metastatic Sites"]
del drug_trial_complete["Mouse ID"]

drug_trial_complete
```

```python
#Because in the directions is does not say anything specific about a list of drugs we are supposed to focus on, I wanted to create a separate array to know the different drug treatment types conducted in this study
list_of_drug_treatments = drug_trial_complete["Drug"].unique()

list_of_drug_treatments
```

```python
#Did a 1)resort of data frame to do a double groupby based on Drug and Timepoint. What I learned is that order matters here. Then 2) I perfromed a calculation to create two new subsets of data for the mean and standard error of the mean (sem). Performed the .unstack to data so I could see what's going across all the timpoints for the different drugs
#I recognize that my data frame below is different from the starter code, but this version is easier for me and to conceptually understand the task. All the information is the same as the starter figure just rearranged differently
mean_and_standardE_TumorV = drug_trial_complete.groupby(["Drug", "Timepoint"]).agg({"Tumor Volume (mm3)" :["mean", "sem"]}).unstack()
mean_and_standardE_TumorV
```

```python
# Established a range of y values for each drug type. Use .loc function tag both the row and specific subset of column data to generate for the list of values established (i used the .tolist()) to perform that. 
#I'm only calling these specific treatments in response towards exemplar that was provided. 
cap_TumorVolMean_Yvalues = mean_and_standardE_TumorV.loc["Capomulin"].loc[:,"mean"].tolist()
inf_TumorVolMean_Yvalues = mean_and_standardE_TumorV.loc["Infubinol"].loc[:,"mean"].tolist()
ket_TumorVolMean_Yvalues = mean_and_standardE_TumorV.loc["Ketapril"].loc[:,"mean"].tolist()
plc_TumorVolMean_Yvalues = mean_and_standardE_TumorV.loc["Placebo"].loc[:, "mean"].tolist()
```

```python
#Same rationale as above, just used standard error of mean (sem) as subset after first .loc filter
cap_TumorVolSEM_Yvalues = mean_and_standardE_TumorV.loc["Capomulin"].loc[:,"sem"].tolist()
inf_TumorVolSEM_Yvalues = mean_and_standardE_TumorV.loc["Infubinol"].loc[:,"sem"].tolist()
ket_TumorVolSEM_Yvalues = mean_and_standardE_TumorV.loc["Ketapril"].loc[:,"sem"].tolist()
plc_TumorVolSEM_Yvalues = mean_and_standardE_TumorV.loc["Placebo"].loc[:,"sem"].tolist()
```

```python
# Establish x_axis for scatter plot. 50 was chosen because the highest domain value was 45. Thus incrementing out this domain by intervals of 5 both made sense and matched the exemplar
x_axis = np.arange(0, 50, 5)
```

```python
#researched this variable to plot all of my error bars on the same figure
fig, ax = plt.subplots(1,1)

#Creation of my different scatter plots that is organized by (x values, y values, sem value, marker shape, color, and label for legend)
plt.errorbar(x_axis, cap_TumorVolMean_Yvalues, yerr=cap_TumorVolSEM_Yvalues, marker= "o", color="red",
                   label="Capomulin")
plt.errorbar(x_axis, inf_TumorVolMean_Yvalues, yerr=inf_TumorVolSEM_Yvalues, marker= "^", color="blue",
                   label="Infubinol")
plt.errorbar(x_axis, ket_TumorVolMean_Yvalues, yerr=ket_TumorVolSEM_Yvalues, marker= "s",color="green",
                   label="Ketapril")
plt.errorbar(x_axis, plc_TumorVolMean_Yvalues, yerr=plc_TumorVolSEM_Yvalues, marker= "d", color="black",
                   label="Placebo")

# Establish all the aesthetic add ons prior to the plt.show() to ensure that they are all on there
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumer Response to Treatment")
plt.legend()
plt.grid(axis="y")
plt.show()

# x lim and y lim not needed because matplotlib already created graph based on best fit and this matches the exemplar. Just wanted to put a note about that
```

```python
# Establish x_axis for scatter plot. Same rationale as previous 2 tasks. 
x_axis = np.arange(0, 50, 5)

#See task 1 for rationale and origin
fig, ax = plt.subplots(1,1)

#Creation of my different scatter plots that is organized by (x values, y values, marker shape, color, and label for legend). The reason why there were no error bars for this specific task was because it would not make sense to determine the sem on a column value that was generated by a .count function 
plt.plot(x_axis, cap_SurvialPercentage_Yvalues, marker= "o", color="red", label="Capomulin")
plt.plot(x_axis, inf_SurvialPercentage_Yvalues, marker= "^", color="blue",label="Infubinol")
plt.plot(x_axis, ket_SurvialPercentage_Yvalues, marker= "s",color="green",label="Ketapril")
plt.plot(x_axis, plc_SurvialPercentage_Yvalues, marker= "d", color="black",label="Placebo")

# Establish all the aesthetic add ons prior to the plt.show() to ensure that they are all on there
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.title("Survival During Treatment")
plt.legend()
plt.grid(axis="y")
plt.show()

# x lim and y lim not needed because matplotlib already created graph based on best fit and this matches the exemplar. Just wanted to put a note about that
```

```python
# Set y values as a defined array that contained the whole number percentages calculated above
y_values = [cap_TumorVolume_change, inf_TumorVolume_change, ket_TumorVolume_change, plc_TumorVolume_change]

# The x_axis need to correlate to the extact length of values in the list above since we are using a bar chart. There should be specific output for each unique input that defines the function used to generate the bar graph
x_axis = np.arange(len(y))


# Needed to create a for loop to append colors to the specific y_values established above. 
bar_graph_colors = []
for item in y_values:
    if item > 0:
        bar_graph_colors.append("red")
    else:
        bar_graph_colors.append("green")


# Created bar graph path that passes the stored x and y values as well as the stored colors.         
percent_TumorVolume_change_graph = plt.bar(x_axis, y_values, color=bar_graph_colors)

# Since the x_axis is not a set of defined integers and we need it to be series of stringed words, I decided to perfrom a list comprehensions to the xtick locations and then passed down a path that we have learned in our classes. 
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Capomulin", "Infubinol", "Ketapril", "Placebo"])


#Other aesthetics for the bar graph 
plt.ylabel("% Tumor Volume Change")
plt.title("Tumor Change Over 45 Day Treatment")
plt.grid()


#The code below does a number of things 1)defines the autolabel function with the shape of the bars to be used in the plot 2)The for loop establishes a series of commands to perform that are dictated by the height of the bars 3)The placement of the percentage value within each bar is included and addressed with .get_width. 4)The aesthetics and alignment of the text within the bars is also included relative to ha, va and color. 
def autolabel(rects):
    for i,rect in enumerate(rects):
        height = int(y_values[i])
        if height >= 0:
            plt.text(rect.get_x()+rect.get_width()/2., 2, "%1.1f%%"% (y_values[i]),
                ha="center", va="bottom", color="white")
        else:
            plt.text(rect.get_x()+rect.get_width()/2., -2, "%1.1f%%"% (y_values[i]),
                ha="center", va="top", color="white")          

autolabel(percent_TumorVolume_change_graph)

plt.show()
```
